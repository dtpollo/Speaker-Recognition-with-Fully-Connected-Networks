### Results

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Resultados: [Accuracy, F1-score, AUC]
Time_R = [0.4787234042553192, 0.47943077891781377, 0.7430084805035845]
Frequency_R = [0.976063829787234, 0.9757112050292184, 0.9956501427988576]
Spectrograms_R = [0.848404255319149, 0.8435141605272912, 0.9911439354199452]
Mel_Spec_R = [0.976063829787234, 0.976122207254309, 0.9996917584659322]
Cepstral_Coeff_R = [0.9707446808510638, 0.970812559591083, 0.9977985224689631]

results = {
    'Tiempo': Time_R,
    'Frecuencia': Frequency_R,
    'Espectrograma': Spectrograms_R,
    'Mel Spectrograma': Mel_Spec_R,
    'MFCC': Cepstral_Coeff_R
}

metrics = ['Accuracy', 'F1 Score', 'AUC']

# Colors
colors = ['#a6cee3', '#b2df8a', '#fb9a99', '#fdbf6f', '#cab2d6']

# Plot
x = np.arange(len(metrics))
width = 0.15

fig, ax = plt.subplots(figsize=(10,6))

for i, (label, values) in enumerate(results.items()):
    ax.bar(x + i*width, values, width, label=label, color=colors[i])

# Configs
ax.set_ylabel('Valor')
ax.set_title('Comparacin de Modelos de Reconocimiento de Oradores')
ax.set_xticks(x + width*2)  # center labels
ax.set_xticklabels(metrics)

# Y steps
ax.set_ylim(0.4, 1.05)
ax.set_yticks(np.arange(0.4, 1.05, 0.05))


ax.grid(True, linestyle='--', alpha=0.7)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```
